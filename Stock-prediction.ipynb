# Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
r = 0
for fig in df.columns:
    es = ExponentialSmoothing(df[fig], trend=None,damped=False, seasonal=None).fit()
    pred = es.predict(start=df[fig].index[0], end=df['OPEN'].index[-1])
    plt.figure(figsize=(15,10))
    
    r += 1
    plt.subplot(2,2,r)
    plt.plot(df[fig].index, df[fig], label='Train')
    plt.plot(pred.index, pred, label='Holt-Winters')
plt.show()
```

# Double Exponential Smoothing

```python
r = 0
for fig in df.columns:
    es = ExponentialSmoothing(df[fig], trend='Add',damped=False, seasonal=None).fit()
    pred = es.predict(start=df[fig].index[0], end=df['OPEN'].index[-1])
    plt.figure(figsize=(15,10))
    
    r += 1
    plt.subplot(2,2,r)
    plt.plot(df[fig].index, df[fig], label='Train')
    plt.plot(pred.index, pred, label='Holt-Winters')
plt.show()
```
