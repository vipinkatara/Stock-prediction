```python
r = 0
for fig in df.columns:    
    r +=1
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,r)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    plt.plot(df[fig])
    plt.xticks([]) , plt.yticks([])
    plt.title(fig)
```

```python
r = 0
for fig in df.columns:    
    r +=1
    plt.figure(figsize=(20,15))
    plt.subplot(2,2,r)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    plt.plot(df[fig])
    plt.plot(df['OPEN'].rolling(5).mean())
    plt.xticks([]) , plt.yticks([])
    plt.title(fig)
    plt.show()
```
